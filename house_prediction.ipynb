### split data


```python
x =df['sqft_living']
y =df['price']
```

```python
# split data
from sklearn.model_selection import train_test_split
x_train , x_test, y_train,y_test = train_test_split(x,y,test_size =0.3,random_state = 101)
```

```python
x_train.shape , x_test.shape
```

### scaling the data 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x_train = scaler.fit_transform(x_train.array.reshape(-1,1))
scaled_x_test = scaler.transform(x_test.array.reshape(-1, 1))
```

## simple Linear Regression

```python
# without scaling
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
# fit training data
reg.fit(x_train.array.reshape(-1, 1) , y_train)

# evaluate the parameter of the model(slop , intercept)
print(reg.intercept_)
print(reg.coef_)
```

```python
# pridict 
y_pred = reg.predict(x_test.array.reshape(-1, 1))
# put them in dataframe
s_df = pd.DataFrame({'Actual': y_test ,'predicted': y_pred})
s_df.head()
```

```python
from sklearn.metrics import mean_squared_error
# calculate loss function using RMSE
print('RMSE : {:.2f}'.format(np.sqrt(mean_squared_error(y_test , y_pred))))
```

```python
print('Linear Regression Model :')
# calculate accuracy of trainig
print('train score  {:.2f}'.format(reg.score(x_train.array.reshape(-1,1),y_train)))
print('test score  {:.2f}'.format(reg.score(x_test.array.reshape(-1,1),y_test)))
```

### LR using Scaling


```python
regressor = LinearRegression()
# fit training data
regressor.fit(scaled_x_train , y_train)

# evaluate the parameter of the model(slop , intercept)
print(regressor.intercept_)
print(regressor.coef_)
```

```python
y_pred = regressor.predict(scaled_x_test)
# put them in dataframe
scaled_df = pd.DataFrame({'Actual': y_test ,'predicted': y_pred})
scaled_df.head()
```

```python
from sklearn.metrics import mean_squared_error
# calculate loss function using RMSE
print('RMSE : {:.2f}'.format(np.sqrt(mean_squared_error(y_test , y_pred))))
```

```python
print('Linear Regression Model :')
# calculate accuracy of trainig
print('train score  {:.2f}'.format(regressor.score(scaled_x_train,y_train)))
print('test score  {:.2f}'.format(regressor.score(scaled_x_test,y_test)))
```

#### scaling does not make any deffirent in the model accuracy
